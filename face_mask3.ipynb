```python
import numpy as np
import cv2
from keras.models import load_model

```

```python
import os
import time
import pyttsx3 as p
engine = p.init()
model = load_model(r'facemask-008.model')

classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

source=cv2.VideoCapture(r"D:\datasets\dataset-mask\mask.mp4")
#source=cv2.VideoCapture(r"D:\datasets\dataset-mask\test_video.mp4")
#source=cv2.VideoCapture(0)
labels_dict={0:'MASK',1:'NO MASK'}
color_dict={0:(0,255,0),1:(0,0,255)}

try:
    if not os.path.exists("D:\datasets\data"):
        os.makedirs("D:\datasets\data")
except OSError:
    print("error while creating directory")
currentframe=0

while(True):
    

    ret,img=source.read()
    
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=classifier.detectMultiScale(gray,1.3,5)  

    for x,y,w,h in faces:
        #cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        face_img=gray[y:y+w,x:x+w]
        resized=cv2.resize(face_img,(100,100))
        normalized=resized/255.0
        reshaped=np.reshape(normalized,(1,100,100,1))
        result=model.predict(reshaped)
        print(result)
        label=np.argmax(result,axis=1)[0]
        
        
        if label == 1:
            
            if currentframe%5==0:
                name=r'D:\datasets\data\frame'+str(currentframe)+'.jpg'
                cv2.imwrite(name,img)
            if currentframe%100==0:
                
                engine.say("wear a mask")
                engine.runAndWait()
            
            
        currentframe+=1
         
        cv2.rectangle(img,(x,y),(x+w,y+h),color_dict[label],2)
        cv2.rectangle(img,(x,y-40),(x+w,y),color_dict[label],-1)
        cv2.putText(img, labels_dict[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
        
        
    cv2.imshow('mask detector',img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cv2.destroyAllWindows()
source.release()
```
